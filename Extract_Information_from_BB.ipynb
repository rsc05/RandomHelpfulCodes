```python
name_of_lectures=[]
html_of_lectures=[]

# Search by text
Units_material=(By.LINK_TEXT,"Work on Progress BB 2023")
Click_Rami(driver, Units_material, 10)
```

```python
# Find all the files in the table
# How many lectures
element2=WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.ID,"listContainer_databody"))).find_elements(By.CSS_SELECTOR, 'tr')
# Get folders in the table
folders_name=[e.find_elements(By.TAG_NAME, 'th')[0].get_attribute("innerText") for e in element2]
folders_name=[e.replace(" ", "") for e in folders_name]
```

```python
html_of_lectures=[]
name_of_lectures=[]
for i in folders_name:
    # click on week
    Units_material=(By.LINK_TEXT,i)
    Click_Rami(driver, Units_material, 10)
    
    # click on lectures
    Units_material=(By.LINK_TEXT,"Lectures")
    Click_Rami(driver, Units_material, 10)
    
    # Open tbody by ID
    element3=WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.ID,"listContainer_databody"))).find_elements(By.CSS_SELECTOR, 'tr')
    name_of_lectures_of_Week=[re.search('(.+?).pdf', e.find_elements(By.TAG_NAME, 'th')[0].get_attribute("innerText")).group(1) for e in element3]
    # Get link
    html_of_lectures_of_Week=[e.find_elements(By.XPATH, ".//descendant::a")[0].get_attribute("href") for e in element3]
    
    # Go back to main page
    driver.back()
    time.sleep(0.1)
    driver.back()
    
    name_of_lectures.extend(name_of_lectures_of_Week)
    html_of_lectures.extend(html_of_lectures_of_Week)
```

```python
d = {'Title of Lecture':name_of_lectures,'HTML of Lecture':html_of_lectures}
df1 = pd.DataFrame(d)   
df1.set_index('Title of Lecture', inplace=True)
df1.head()
```
